```python
fa = pd.read_csv('sfa1314.csv')
fa.head()
```

```python
gr = pd.read_csv('gr200_14.csv')
gr.head()
```

```python
fa.columns
```

```python
hd = pd.read_csv('hd2015.csv',encoding="ISO-8859-1")
```

```python
hd.head()
```

```python
data = pd.merge(hd, fa, left_on='UNITID', right_on='UNITID')
```

```python
data.head()
```

```python
data = pd.merge(data, gr, left_on='UNITID', right_on='UNITID')
```

```python
datadf = data[['INSTNM','PGRNT_P','BAGR100','NPIS422','NPIS432']]
```

INSTNM = College name
1. PGRNT_P = % of first year students receiving pell grant
SCUGFFN = Total number of full-time first-time degree/certificate seeking undergraduates - financial aid cohort
2. BAGR100 = 4-year Graduation rate - bachelor's degree within 100% of normal time
3. NPIS422 = Average net price (income 30,001-48,000)-students receiving Title IV Federal financial aid, 2013-14
4. NPIS432 = Average net price (income 48,001-75,000)-students receiving Title IV Federal financial aid, 2013-14

```python
datadf.head()
```

```python
datadf['index1'] = datadf['PGRNT_P'].multiply(datadf['BAGR100'], axis="index")
```

```python
datadf.head()
```

```python
datadf['index2'] = datadf['NPIS422'] + datadf['NPIS432'] / 2
```

```python
datadf.dropna()
```

```python
datadf.head()
```

```python
datadf['finalindex'] = (datadf['index1']/datadf['index1'].mean()) + (datadf['index2']/datadf['index2'].mean())
```

```python
datadf = datadf.dropna()
```

```python
datadf.sort_values('finalindex', ascending=False)
```
